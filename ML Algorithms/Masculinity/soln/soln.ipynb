```python
df1 = df[['q0001', 'q0002','q0007_0001', 'q0007_0002', 'q0007_0003', 'q0007_0004',
       'q0007_0005', 'q0007_0006', 'q0007_0007', 'q0007_0008', 'q0007_0009',
       'q0007_0010', 'q0007_0011','q0009','q0024','q0025_0002','age3', 'kids','weight']]
       #Removing all the unnecessary columns from the dataframe after  hovering all of the things ,kept the things which are strong for masculinity 
```

```python
df1.dropna(inplace = True)          #Removing all the null values directly as they are very less in comparision to the shape of the dataframe
df1.head()
```

```python
df1.q0024.unique()      # All kinds of values in q0024
```

```python
df1['q0025_0002'][df1['q0024'] == 'Never married'].unique() #  Finding all the values of q0025_0002 where q0024 is  == to Never married
```

```python
temp = df1[df1['q0025_0002'] == 'Yes, one or more children 18 or older']        # storing all the rows where 'q0025_0002' == 'Yes, one or more children 18 or older in temp'
df2 = df1[df1['q0025_0002'] !='Yes, one or more children 18 or older']      #Removing all the rows where  'q0025_0002'  not == 'Yes, one or more children 18 or older'
df2 = pd.concat([df2,temp[temp['age3'] != '18 - 34']],ignore_index=True)    # removing all the values of temp where age == 18- 34 and storing the remaning values to the orignal DataFrame
df2.head()
# Basiclly this process was done to remove the rows where the fathers age is between 18- 34 where the sons age is greater then or equal to 18
# because the fathers age cannot be 18 - 34 with the sons age being above or equal to 18
#  eg if father was married at 18(standard age of marrage is 21) and had kids in the same year.
# So when the so sons age will get 18 the fathers age should be 18 +18 = 36 which does not fall in (18 - 34)
```

```python
def age(df):
    if 'up' not in df:
        temp = np.random.randint(int(df.split(' ')[0]),int(df.split(' ')[-1]))
        return temp
    return np.random.randint(int(df.split(' ')[0]),80)
df2['age3'] = df2['age3'].apply(age)            #Encrypting the string data type to int in age column
df2.head()
```

```python
df2.drop(['q0024','q0025_0002','kids'],axis = 'columns',inplace = True) # Droping the columns as their use is over 
df2.head()
```

```python
def colum(df):
    if df!= 'No answer':
        if len(df.split(' ')) > 1:
            if 'Never' in df:
                if 'not' in df:
                    return 1
                return 2
        if df.strip() == 'often':
                return 4
        if df.strip() == 'Rarely':
                return 3
    return 0

df2['q0007'] = 0
for column in df2.columns[2:-4]:
    df2[column] = df2[column].apply(colum)      #Encrypting the string data type to int in the q0007 series column
    df2['q0007'] += df2[column]                 # Adding them up

df2.drop(['q0007_0001', 'q0007_0002', 'q0007_0003', 'q0007_0004', 'q0007_0005', 'q0007_0006', 'q0007_0007', 'q0007_0008', 'q0007_0009', 'q0007_0010', 'q0007_0011'],axis = 'columns', inplace = True)
df2
        
```

```python
df2.q0009.unique()  #Finding all the values in q0009
```

```python
df2.q0007.unique()  #Finding all the values in q0007
```

```python
def column(df):
    if df != 'No answer':
        if df == 'Not employed, NOT looking for work':
            return -1
        if df == 'Not employed, student':
            return 0
        if df == 'Not employed, looking for work':
            return 1
        if df == 'employed, working part-time':
            return 2
        if df == 'Employed, working full-time':
            return 3
    return 0
        
df2['q0009'] = df2['q0009'].apply(column)
df2['Status'] = df2['q0009'] + df2['q0007']     # Adding all the values to the status column which shows the status of masuculinity
df2.drop(['q0007','q0009'],axis = 'columns',inplace = True)     #Droping the unnecessary rows
df2.head()
```

```python
df2['q0001'].unique()       #Finding all the values in q0001
```

```python
def converter(df):
    if 'Somewhat' in df:
        return 3
    if 'Not very' in df:
        return 2
    if 'Not at all' in df:
        return 1
    if "Very" in df:
        return 4
    else:
        return 0
df2['q0001'] = df2.q0001.apply(converter)#Encrypting the string data type to int in q0001 column
df2['q0002'] = df2.q0002.apply(converter)#Encrypting the string data type to int in q0002 column
df2.head()
```

```python
plt.scatter(df2.age3,df2.Status)        #plot of age vs status 
```

```python
from sklearn.cluster import KMeans
ssd  =[]
for k in range(1,11):
    km = KMeans(k)
    y = km.fit_predict(df2[['Status','age3']])
    ssd.append(km.inertia_)
plt.plot(list(range(1,11)),ssd)
plt.show()
plt.clf()           # Number of cluster required for efficient clustring
```

```python
km = KMeans(n_clusters = 2)
y_predicted = km.fit_predict(df2[['Status','age3']])        # Predicted output of given dataset
df2 = pd.concat([df2,pd.DataFrame(y_predicted,columns=['Y_predicted'])],axis = 1)       #concatenated the output to the orignal dataframe 
df2.head()
```

```python
plt.scatter(df2[df2['Y_predicted'] == 0]['age3'],df2[df2['Y_predicted'] == 0]['Status'],color = 'red',label = 'young')
plt.scatter(df2[df2['Y_predicted'] == 1]['age3'],df2[df2['Y_predicted'] == 1]['Status'],color = 'green',label = 'old')
plt.xlabel('Age')
plt.ylabel('Masculinity')
plt.legend()
plt.show()
plt.clf()
#Graph of the age vs status(masculinity) we can see that as age increases the green part gets concentrated at the bottum part of the graph 
# which means that as age increases masculinity decreases
# where as for young it is evenly spreaded but slightly concentrated at the upper part of the graph. 
```
